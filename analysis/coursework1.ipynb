```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("a - length.txt", sep="\t", header=None,names=["start_len","end_len","count"])
```

```python
data['size'] = data['start_len'].map(str) + '-' + data['end_len'].map(str)
data
```

```python
data.plot.bar(figsize=(20,10), x='size', y='count', title="Count tweet length", fontsize=16)
```

```python
filtered_data = data[data["start_len"] <= 155]
```

```python
filtered_data.plot.bar(figsize=(20,10), x='size', y='count', title="Count tweet length", fontsize=16)
```
